```python
#導入所需套件

import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from keras.models import Sequential
from keras import layers
from keras.optimizers import *
from sklearn.metrics import roc_auc_score
```

```python
#讀進訓練集和測試集

train_df = pd.read_csv('train/train.csv',encoding='Big5')
test_df = pd.read_csv('test/test.csv',encoding='Big5')
train_df.head()
```

```python
#移除不需要的特徵

train_df=train_df.drop(columns=['LAST_B_ISSUE_DT','IF_Y_REAL_IND','IF_ISSUE_O_IND','IF_ISSUE_M_IND','CUS_ID'])
test_df=test_df.drop(columns=['LAST_B_ISSUE_DT','IF_Y_REAL_IND','IF_ISSUE_O_IND','IF_ISSUE_M_IND','CUS_ID'])
train_df.head()
```

```python
#設定標籤值

cnt=0
for i in train_df['Y1']:
    if i=='N':
        train_df['Y1'][cnt]=0
    elif i=='Y':
        train_df['Y1'][cnt]=1
    cnt+=1

label=train_df['Y1']
label=np.array(label, dtype=int)
train_df=train_df.drop(columns=['Y1'])
train_df.head()
```

```python
#對特徵進行優化處理

#訓練集
train_df['OCCUPATION_CLASS_CD']=train_df['OCCUPATION_CLASS_CD'].fillna(6)
train_df=pd.get_dummies(data=train_df,columns=['OCCUPATION_CLASS_CD'])
train_df['TERMINATION_RATE']=train_df['TERMINATION_RATE'].fillna(train_df['TERMINATION_RATE'].max())
train_df['EXPIRATION_AMT']=train_df['EXPIRATION_AMT'].fillna(train_df['EXPIRATION_AMT'].max())
train_df['ILL_ADDITIONAL_AMT']=train_df['ILL_ADDITIONAL_AMT'].fillna(train_df['ILL_ADDITIONAL_AMT'].max())
train_df['LONG_TERM_CARE_AMT']=train_df['LONG_TERM_CARE_AMT'].fillna(train_df['LONG_TERM_CARE_AMT'].max())
train_df['APC_1ST_YEARDIF']=train_df['APC_1ST_YEARDIF'].fillna(train_df['APC_1ST_YEARDIF'].mean())
train_df['POLICY_VALUE_AMT']=train_df['POLICY_VALUE_AMT'].fillna(train_df['POLICY_VALUE_AMT'].max())
train_df['ILL_ACCELERATION_AMT']=train_df['ILL_ACCELERATION_AMT'].fillna(train_df['ILL_ACCELERATION_AMT'].max())
train_df['ANNUITY_AMT']=train_df['ANNUITY_AMT'].fillna(train_df['ANNUITY_AMT'].max())
train_df['FIRST_CANCER_AMT']=train_df['FIRST_CANCER_AMT'].fillna(train_df['FIRST_CANCER_AMT'].max())
train_df['OUTPATIENT_SURGERY_AMT']=train_df['OUTPATIENT_SURGERY_AMT'].fillna(train_df['OUTPATIENT_SURGERY_AMT'].max())
train_df['BMI']=train_df['BMI'].fillna(train_df['BMI'].mean())
train_df['ANNUAL_INCOME_AMT']=train_df['ANNUAL_INCOME_AMT'].fillna(train_df['ANNUAL_INCOME_AMT'].max())

#測試集
test_df['OCCUPATION_CLASS_CD']=test_df['OCCUPATION_CLASS_CD'].fillna(6)
test_df=pd.get_dummies(data=test_df,columns=['OCCUPATION_CLASS_CD'])
test_df['TERMINATION_RATE']=test_df['TERMINATION_RATE'].fillna(test_df['TERMINATION_RATE'].max())
test_df['EXPIRATION_AMT']=test_df['EXPIRATION_AMT'].fillna(test_df['EXPIRATION_AMT'].max())
test_df['ILL_ADDITIONAL_AMT']=test_df['ILL_ADDITIONAL_AMT'].fillna(test_df['ILL_ADDITIONAL_AMT'].max())
test_df['LONG_TERM_CARE_AMT']=test_df['LONG_TERM_CARE_AMT'].fillna(test_df['LONG_TERM_CARE_AMT'].max())
test_df['APC_1ST_YEARDIF']=test_df['APC_1ST_YEARDIF'].fillna(test_df['APC_1ST_YEARDIF'].mean())
test_df['POLICY_VALUE_AMT']=test_df['POLICY_VALUE_AMT'].fillna(test_df['POLICY_VALUE_AMT'].max())
test_df['ILL_ACCELERATION_AMT']=test_df['ILL_ACCELERATION_AMT'].fillna(test_df['ILL_ACCELERATION_AMT'].max())
test_df['ANNUITY_AMT']=test_df['ANNUITY_AMT'].fillna(test_df['ANNUITY_AMT'].max())
test_df['FIRST_CANCER_AMT']=test_df['FIRST_CANCER_AMT'].fillna(test_df['FIRST_CANCER_AMT'].max())
test_df['OUTPATIENT_SURGERY_AMT']=train_df['OUTPATIENT_SURGERY_AMT'].fillna(test_df['OUTPATIENT_SURGERY_AMT'].max())
test_df['BMI']=train_df['BMI'].fillna(test_df['BMI'].mean())
test_df['ANNUAL_INCOME_AMT']=test_df['ANNUAL_INCOME_AMT'].fillna(test_df['ANNUAL_INCOME_AMT'].max())

train_df.head()
```

```python
#對資料進行補值

train_df=train_df.fillna(0)
test_df=test_df.fillna(0)
train_df.head()
```

```python
#對資料進行one-hot encoding

#訓練集
cnt_list=list()
df_name=list(train_df)
column_name=list()
cnt=0
for i in train_df.iloc[629,:]:
    if type(i)==str:
        cnt_list.append(cnt)
    cnt+=1
for i in cnt_list:
    column_name.append(df_name[int(i)])
# len(column_name)
    train_extend=pd.get_dummies(data=train_df,columns=column_name)

#測試集
cnt_list=list()
df_name=list(test_df)
column_name=list()
cnt=0
for i in test_df.iloc[1543,:]:
    if type(i)==str:
        cnt_list.append(cnt)
    cnt+=1
for i in cnt_list:
    column_name.append(df_name[int(i)])
# len(column_name)
    test_extend=pd.get_dummies(data=test_df,columns=column_name)
    
train_extend.head()
```

```python
train_feature=train_extend.values
test_feature=test_extend.values

minmax_scale=preprocessing.MinMaxScaler(feature_range=(0,1))
train_minmax_feature=minmax_scale.fit_transform(train_feature)
test_minmax_feature=minmax_scale.fit_transform(test_feature)

print("Train data normalization:",train_minmax_feature[0])
print("Test data normalization:",test_minmax_feature[0])
```

```python
#將訓練集做切割(訓練集和驗證集)

x,y=shuffle(train_minmax_feature,label,random_state=2)
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.1,random_state=10)
```

```python
#建立訓練模型

model = Sequential()
model.add(layers.Dense(1024, activation='relu',input_shape=(292,)))
model.add(layers.Dropout(0.4))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(optimizer=Adam(1e-3,decay=1e-4), loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#開始訓練模型

model.fit(x,y,validation_data=(test_x,test_y),epochs=6,batch_size=100,verbose=1)
predict1=model.predict(test_x)
predict2=model.predict(train_x)
score1=roc_auc_score(test_y, predict1)
score2=roc_auc_score(train_y, predict2)
print("Train data score:",score1)
print("Test data score:",score2)
```

```python
#預測測試集並輸出預測檔案

predict=model.predict(test_minmax_feature)
sub=pd.read_csv('submit_test/submit_test.csv',sep=',')
sub['Ypred']=predict
sub.to_csv('predict1.csv',index=False)
```
